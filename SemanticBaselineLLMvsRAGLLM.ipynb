```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python

sns.set_theme(style='white',font='DejaVu Sans')
pairs=[('TinyLlama','RAG TinyLlama'),('Mistral-7B','RAG Mistral-7B'),('Llama-3.1 8B','RAG Llama-3.1 8B'),('Llama-13B','RAG Llama-13B')]
recall={'TinyLlama':0.857397,'RAG TinyLlama':0.876795,'Mistral-7B':0.8712851,'RAG Mistral-7B':0.894786,'Llama-3.1 8B':0.8705658,'RAG Llama-3.1 8B':0.885985,'Llama-13B':0.855844,'RAG Llama-13B':0.861476}
f1={'TinyLlama':0.824225,'RAG TinyLlama':0.8388122,'Mistral-7B':0.8514198,'RAG Mistral-7B':0.875506,'Llama-3.1 8B':0.819228,'RAG Llama-3.1 8B':0.8291933,'Llama-13B':0.8073012,'RAG Llama-13B':0.8102933}
blue='#003366'; orange='#A33C00'; width=0.35
fig,axes=plt.subplots(1,2,figsize=(12,5))
for ax,metric,d in zip(axes,['BERT Recall','BERT F1 Score'],[recall,f1]):
    for idx,(base,rag) in enumerate(pairs):
        ax.bar(idx-width/2,d[base],width,color=blue,label='Baseline' if idx==0 else "")
        ax.bar(idx+width/2,d[rag],width,color=orange,label='RAG' if idx==0 else "")
    ax.set_xticks(range(len(pairs)))
    ax.set_xticklabels([b.split()[0] for b,_ in pairs])
    ax.set_ylabel('Score')
    ax.set_title(metric)
    ax.set_ylim(0,1)
    ax.legend(loc='upper right',frameon=False)
plt.tight_layout()
plt.savefig('BERT_Recall_F1_grouped_v3.png',dpi=300)
```
